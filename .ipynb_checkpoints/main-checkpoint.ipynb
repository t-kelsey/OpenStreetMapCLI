```python
import pyrosm

address = "Lehener Str. 90 Freiburg"
radius = 3 # in km
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent='myapplication')
location = geolocator.geocode(address)

print(location.address, location.longitude, location.latitude)
```

```python
import math

fp = pyrosm.get_data("germany")

    # Latitude: 1 deg = 110.574 km
    # Longitude: 1 deg = 111.320*cos(latitude) km

latitude_radius = float(radius) / 110.574
longitude_radius = float(radius) / (111.320 * math.cos(math.radians(latitude_radius)))
bounding_box = [location.longitude - longitude_radius, location.latitude - latitude_radius, location.longitude + longitude_radius, location.latitude + latitude_radius]
print(bounding_box)

osm = pyrosm.OSM(fp, bounding_box=bounding_box)
```

```python
import re

custom_filter = {'healthcare' : True}

pois = osm.get_pois(custom_filter=custom_filter, extra_attributes=["healthcare"])
pois["poi_type"] = pois["healthcare"]

buildings = osm.get_buildings()

housenumber_regex_match = re.search(r"[\w\- ]*", location.address)
housenumber = housenumber_regex_match.group(0)

street_regex_match = re.search(r",[\wßüöä\- ]*", location.address)
street = street_regex_match.group(0)[2:]
print(street, housenumber)

user_location = buildings[(buildings["addr:street"] == street) & (buildings["addr:housenumber"] == housenumber)]
print("user location: ", user_location)

ax = buildings.plot()
pois.plot(column="poi_type", ax=ax, legend=True, markersize=2, legend_kwds=dict(loc='upper left', ncol=1, bbox_to_anchor=(1, 1)))
```
